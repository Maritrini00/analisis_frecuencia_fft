```python
"""
Hacer el análisis de frecuencia de 5 señales diferentes (voz, música, sonido ambiente, bioseñal, etc.). Usar el comando np.fft.fft de python.
a) Graficar espectro de magnitud.
b) Graficar espectro de fase.
 """
```

```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
def analisis_freq(file):
  fs, data = wavfile.read(file)

  #señal
  if len(data.shape) == 2:
    signal = data[:, 0]  # Consideramos solo un canal de audio (mono)
  else:
    signal = data
  #tamaño senal
  N = len(signal)
  #vector de tiempos
  t = np.arange(0,N)/fs

  #análisis de frecuencia
  freq = np.fft.fftfreq(N,1/fs)
  result_fft = np.fft.fft(signal)

  pos_mask = freq >= 0
  freqs_pos = freq[pos_mask]
  fft_pos = result_fft[pos_mask]

  #Gráfico de la señal
  plt.figure(1)
  plt.plot(signal)
  plt.title('Señal de - ' + file)

  plt.show()

  #espectro de magnitud
  plt.figure(2)
  plt.plot(freqs_pos, np.abs(fft_pos))
  plt.title('Espectro de Magnitud - ' + file)
  plt.xlabel('Frecuencia (Hz)')
  plt.ylabel('Magnitud')
  plt.show()

  #espectro de fase
  plt.figure(3)
  plt.plot(freqs_pos, np.angle(fft_pos))
  plt.title('Espectro de Fase - ' + file)
  plt.xlabel('Frecuencia (Hz)')
  plt.ylabel('Fase')
  plt.show()
```

```python
cat_meow = 'cat_meow_x.wav'
analisis_freq(cat_meow)
```

```python
dentist = 'dentist_drill.wav'
analisis_freq(dentist)
```

```python
pacman = 'pacman_intro.wav'
analisis_freq(pacman)
```

```python
frog = 'frog.wav'
analisis_freq(frog)
```
